### Database structure

- shops
  - shop_id : INTEGER PK ~ "shop_index"
  - domain_url : VARCHAR (NOTNULL) ~ "Online shop URL"
  - domain_registration_date : DATETIME (NULLABLE) ~ "Domain registration date"
  - is_young_domain : BOOL (NOTNULL) ~ "Indication of young domain"
  - contact_email : BOOL (NOTNULL) ~ "Presence of free contact emails"
  - logo_url : BOOL (NOTNULL) ~ "Presence of logo URL"
  - legitimate : BOOL (NOTNULL) ~ "Label"

- payment_methods
  - payment_method_id : INTEGER PK
  - allow_credit_card : BOOL (NOTNULL) ~ "Presence of credit card payment"
  - allow_money_back : BOOL (NOTNULL) ~ "Presence of money back payment"
  - allow_cash_on_delivery : BOOL (NOTNULL) ~ "Presence of cash on delivery payment"
  - allow_crypto : BOOL (NOTNULL) ~ "Presence of crypto currency"
  - shop_id : INTEGER FK

- certs
  - cert_id : INTEGER PK
  - cert_expire_date : DATE (NOTNULL) ~ "SSL certificate expire date"
  - shop_id : INTEGER FK (NOTNULL)
  - ssl_id : INTEGER FK (NOTNULL)

- ssl_issuer
  - ssl_id : INTEGER PK 
  - ssl_issuer : VARCHAR (NOTNULL) ~ "SSL certificate issuer"
  - ssl_organization : VARCHAR (NOTNULL) ~ "Issuer organization"

- review
  - review_id : INTEGER PK
  - trust_pilot_score : INTEGER (NULLABLE) ~ "TrustPilot score"
  - site_jabber_review : BOOL (NULLABLE) ~ "Presence of SiteJabber reviews"
  - tranco_rank : INTEGER (NULLABLE) ~ "Presence in the standard Tranco list"
  - shop_id : INTEGER FK

```python
shops = df[[
    "shop_id",
    "Online shop URL",
    "Domain registration date",
    "Indication of young domain ",
    "Presence of free contact emails",
    "Presence of logo URL",
    "Label"
]].rename(columns={
    "shop_id": "shop_id",
    "Online shop URL": "domain_url",
    "Domain registration date": "domain_registration_date",
    "Indication of young domain ": "is_young_domain",
    "Presence of free contact emails": "contact_email",
    "Presence of logo URL": "logo_url",
    "Label": "legitimate"
})

shops["legitimate"] = shops["legitimate"].map({"fraudulent": 0, "legitimate": 1})

def registration_to_datetime(ss):
    
    try:
        date = datetime.datetime.strptime(ss, "%Y-%m-%d %H:%M")
    except:
        return None
    
    date_str = date.strftime("%Y-%m-%d %H:%M:%S")
    return date_str

shops["domain_registration_date"] = shops["domain_registration_date"].apply(registration_to_datetime)
```

```python
payment_methods = df[[
    "Presence of credit card payment",
    "Presence of money back payment",
    "Presence of cash on delivery payment",
    "Presence of crypto currency",
    "shop_id"
]].rename(columns={
    "shop_id": "shop_id",
    "Presence of credit card payment": "allow_credit_card",
    "Presence of money back payment": "allow_money_back",
    "Presence of cash on delivery payment": "allow_cash_on_delivery",
    "Presence of crypto currency": "allow_crypto"
})
payment_methods["payment_method_id"] = payment_methods.index

review = df[[
    "TrustPilot score",
    "Presence of SiteJabber reviews",
    "Tranco List rank",
    "shop_id"
]].rename(columns={
    "shop_id": "shop_id",
    "TrustPilot score": "trust_pilot_score",
    "Presence of SiteJabber reviews": "site_jabber_review",
    "Tranco List rank": "tranco_rank"
})

review["review_id"] = review.index
```

```python
ssl_issuer = df[["Issuer organization", "SSL certificate issuer"]]
ssl_issuer = ssl_issuer.drop_duplicates(keep='first').reset_index(drop=True)
ssl_issuer["ssl_id"] = ssl_issuer.index
ssl_issuer = ssl_issuer.rename(columns={
    "Issuer organization": "ssl_organization",
    "SSL certificate issuer": "ssl_issuer"
})
```

```python
certs = df[[
    
    'SSL certificate issuer',
    'SSL certificate expire date',
    'Issuer organization',
    'shop_id', 
]].rename(columns={
    "Issuer organization": "ssl_organization",
    "SSL certificate issuer": "ssl_issuer",
    "SSL certificate expire date": "cert_expire_date"
})

certs = ssl_issuer.merge(certs, on=['ssl_issuer', 'ssl_organization'], how='left')[["shop_id", "ssl_id", "cert_expire_date"]]
certs = certs.sort_values(by='shop_id', ascending=True).reset_index(drop=True)
certs = certs.rename(columns={
    "shop_id": "shop_id_fk",
    "ssl_id": "ssl_id_fk"
})

certs["cert_id"] = certs.index

# Format date

import datetime

def expire_to_datetime(ss):
    try:
        date = datetime.datetime.strptime(ss, "%b %d %H:%M:%S %Y GMT")
    except:
        return None
    date_str = date.strftime("%Y-%m-%d %H:%M:%S")
    return date_str

certs["cert_expire_date"] = certs["cert_expire_date"].apply(expire_to_datetime)
```
